```python
from skimage import color

imageGray = color.rgb2gray(image)
io.imshow(imageGray)
```

```python
plt.contour(imageGray, origin = "image")  #origin = "image", else inverted
```

```python
from skimage.transform import rescale
image_rescaled = rescale(imageGray, 0.75)
io.imshow(image_rescaled)
```

```python
from skimage import filters
edges = filters.sobel(imageGray)
io.imshow(edges)
io.show()
```

```python
import numpy as np

f = np.fft.fft2(imageGray)
# Bring the zero-frequency component to the center
fshift = np.fft.fftshift(f)
magnitude_spectrum = np.log(np.abs(fshift))

plt.imshow(imageGray, cmap = "gray")
```

```python
plt.imshow(magnitude_spectrum)
plt.show()
```

```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt

camera = data.camera()
print(camera)
io.imshow(camera)
```

```python
camera[:30,:] = 0
camera[:,:30] = 0

plt.imshow(camera, cmap = 'gray')
plt.show()
```

```python
mask = camera < 87
camera[mask] = 255

plt.imshow(camera, cmap = 'gray')
plt.show()
```

```python
l_x, l_y = camera.shape[0], camera.shape[1]
print(l_x, l_y)
X, Y = np.ogrid[:l_x, :l_y]
print(X,Y)
outer_disk_mask = (X - l_x / 2)**2 + (Y - l_y / 2)**2 > (l_x / 2)**2
camera[outer_disk_mask] = 0

plt.imshow(camera, cmap = 'gray')
plt.show()
```
